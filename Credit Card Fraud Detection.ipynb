# Importing libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

#### Loading dataset 

```python
credit_card_data=pd.read_csv("C:/Users/LENOVO/Desktop/AI/Credit card fraud detection/creditcard.csv")
```

```python
#Printing first five dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#dataset information
credit_card_data.info()
```

```python
#check missing values
credit_card_data.isnull().sum()
```

```python
#distribution of legit transaction and fraudlent transaction
credit_card_data["Class"].value_counts()
```

```python
#Separating the data for analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Compare the values for both transactions
credit_card_data.groupby("Class").mean()
```

```python
#Undersampling
#Build a sample dataset containing similar distribution of normal transactions and fraudulent transactions
#No. of fraudulent transactions =492
legit_sample=legit.sample(n=492) 
```

```python
#Concatenating two dataframes
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset["Class"].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Splitting the data into features and targets
X=new_dataset.drop(columns="Class",axis=1)
Y=new_dataset["Class"]
```

```python
print(X)
```

```python
print(Y)
```

```python
#Split the data into training data and testing data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

```python
#Model training-Logistic regression
model=LogisticRegression()
```

```python
#Training the logistic regression model with training data
model.fit(X_train,Y_train)
```

```python
#Model evaluation
#Accuracy score
#Accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy of training data is: ",training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy of test data is : ',test_data_accuracy)
```
